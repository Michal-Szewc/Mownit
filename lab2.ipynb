```python
import matplotlib.pyplot as plt
import math
```

```python
def mul(x,X,i):
    wyn = 1
    for j in range(len(X)):
        if j is not i:
            wyn *= (x - X[j])
    return wyn

def L(X,Y,x):
    wyn = 0
    for i in range(len(X)):
        wyn += Y[i]*mul(x,X,i)/mul(X[i],X,i)
    return wyn
```

```python
def r(X,Y,l,p,pom):
    if pom[l][p]:
        return pom[l][p]
    if l == p:
        pom[l][p] = Y[l]
    else:
        pom[l][p] = (r(X,Y,l+1,p,pom) - r(X,Y,l,p-1,pom))/(X[p] - X[l])
    return pom[l][p]

def mulN(X,k,x):
    il = 1
    for j in range(k):
        il *= x - X[j]
    return il
        
def N(X,Y,x,pom):
    wyn = 0
    for i in range(len(Y)):
        wyn += r(X,Y,0,i,pom) * mulN(X,i,x)
    return wyn
```

```python
def czeb(k):
    return [math.cos((2*j -1)/(2*k) * math.pi) for j in range(1,k+1)]
```

```python
def f(x):
    return math.sin(x)
```

```python
X = list(range(-10,11))
Y = [f(x) for x in X]
X_L = [x/100 for x in range(-300,300)]
Y_L = [L(X,Y,x) for x in X_L]

plt.plot(X_L,Y_L)
```

```python
X = czeb(20)
Y = [f(x) for x in X]
X_L = [x/100 for x in range(-300,300)]
Y_L = [L(X,Y,x) for x in X_L]

plt.plot(X_L,Y_L)
```

```python
X = [x/10 for x in range(-10,11)]
pom = [[None for _ in X] for _ in X]
Y = [f(x) for x in X]
X_N = [x/100 for x in range(-300,300)]
Y_N = [N(X,Y,x,pom) for x in X_N]

plt.plot(X_N,Y_N)
```
