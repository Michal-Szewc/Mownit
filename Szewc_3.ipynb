```python
import matplotlib.pyplot as plt
import math
```

```python
def mul(x,X,i):
    wyn = 1
    for j in range(len(X)):
        if j is not i:
            wyn *= (x - X[j])/(X[i] - X[j])
    return wyn

def L(X,Y,x):
    wyn = 0
    for i in range(len(X)):
        wyn += Y[i]*mul(x,X,i)
    return wyn
```

```python
def sklejane(X,Y,x,k):
    y = []
    index =0
    for i in range(len(X)-k):
        while index < len(x) and x[index] <= X[i + k]:
            y += [L(X[i:(i + k + 1)],Y[i:(i + k + 1)],x[index])]
            index += 1
    return y
```

```python
def czeb(k):
    return [math.cos((2*j -1)/(2*k) * math.pi) for j in range(1,k+1)]
```

```python
def reg(k): # k rowno od 0 do 1
    return [x/(k - 1) for x in range(k)]
```

```python
def scale(x,l,p): # z [o,1] a [l,p]
    return x * (p-l) + l
```

```python
def f(x):
    k = 1
    m = 1
    return math.sin(m*x)*math.sin(k*x**2/math.pi)
```

```python
def diff(Y_1,Y_2):
    diff=0
    for i in range(len(Y_2)):
        diff = max(diff,abs(Y_1[i]-Y_2[i]))
    return diff

def error(Y_1,Y_2):
    error = 0
    for i in range(len(Y_2)):
        error += (Y_1[i] - Y_2[i]) ** 2
    return error/len(Y_1)
```

```python
points = 10
n = 1000
```

```python
f_X = [scale(x,(-2) * math.pi,math.pi) for x in reg(n)]
f_Y = [f(x) for x in f_X]

plt.plot(f_X,f_Y)
```

```python
X = [scale(x,(-2) * math.pi,math.pi) for x in reg(points)]
Y = [f(x) for x in X]
X_S = [scale(x,(-2) * math.pi,math.pi) for x in reg(n)]
Y_S = sklejane(X,Y,X_S,0)

print("Max difference: ")
print(diff(f_Y,Y_S))
print("standard error: ")
print(error(f_Y,Y_S))

plt.plot(f_X,f_Y,'r-')
plt.plot(X_S,Y_S)
plt.plot(X,Y,'o')
```
